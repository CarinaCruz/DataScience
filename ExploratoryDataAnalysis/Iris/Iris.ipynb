## Carregando o Conjunto de dados do SKLEARN 

```python
# Imports
import time
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
# Carregando o dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
data.head()
```

## Extração e Transformação de Dados

```python
# visualizando os nomes das colunas 
iris.target_names
```

```python
# visualizando os valores numéricos da variável target
iris.target
```

```python
# Adicionando ao dataset uma nova coluna com os nomes das espécies
data['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
data.head(10)
```

```python
# Incluindo no dataset uma coluna com os valores numéricos da variável target
data['num_target'] = iris.target
data.head(10)
```

```python
# média de cada feature para as três classes target
data.groupby('num_target').mean().T
```
