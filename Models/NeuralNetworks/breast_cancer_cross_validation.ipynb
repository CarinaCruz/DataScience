# <b> Problema de Classificação Binária - Breast Cancer com Cross Validation

```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
previsores = pd.read_csv('entradas.csv')
classe = pd.read_csv('saidas.csv')
```

```python
# Dropout para prevenir overfitting (zerando 20% da camada dos neurônios de entrada e na oculta)
def criarRede():
    classificador = Sequential()
    classificador.add(Dense(units = 16, activation = 'relu', 
                        kernel_initializer = 'random_uniform', input_dim = 30))
    classificador.add(Dropout(0.2))
    classificador.add(Dense(units = 16, activation = 'relu', 
                        kernel_initializer = 'random_uniform'))
    classificador.add(Dropout(0.2))
    classificador.add(Dense(units = 1, activation = 'sigmoid'))
    otimizador = keras.optimizers.Adam(lr = 0.001, decay = 0.0001, clipvalue = 0.5)
    classificador.compile(optimizer = otimizador, loss = 'binary_crossentropy',
                      metrics = ['binary_accuracy'])
    return classificador
```

```python
# Criação da Rede Neural
classificador = KerasClassifier(build_fn = criarRede,
                                epochs = 100,
                                batch_size = 10)
```

```python
# Validando com Cross Validation com 10 rodadas
resultados = cross_val_score(estimator = classificador,
                             X = previsores, y = classe,
                             cv = 10, scoring = 'accuracy')
```

```python
# Visualizando as métricas da avaliação
media = resultados.mean()

# quanto maior mais tendência à overfitting
desvio = resultados.std() 
```
