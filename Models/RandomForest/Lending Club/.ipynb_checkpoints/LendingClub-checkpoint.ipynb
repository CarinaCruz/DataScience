    Utilizaremos os dados de empréstimos de 2007-2010 e tentaremos
    classificar e prever se o mutuário pagou o empréstimo na íntegra.

## <b>Importando bibliotecas

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## <b>Importando o dataset

```python
loans = pd.read_csv('loan_data.csv')
```

## <b> Visualizando as informações do dataset

```python
loans.info()
```

```python
loans.describe()
```

```python
loans.head()
```

## <b> Análise Exploratória dos Dados

```python
# Visualizando a relação entre o FICO (pontuação de crédito) e credit.policy 
# (se o cliente atender aos critérios de subscrição de crédito)
sns.set_style('whitegrid')
plt.figure(figsize = (10,6))
loans[loans['credit.policy'] == 1]['fico'].hist(alpha = 0.5,color = 'blue', 
                                              bins = 30,label = 'Credit.Policy=1')
loans[loans['credit.policy'] == 0]['fico'].hist(alpha = 0.5,color = 'red', 
                                              bins = 30,label = 'Credit.Policy=0')
plt.legend()
plt.xlabel('FICO')
```

```python
# Visualizando a relação entre o FICO (pontuação de crédito) e not.fully.paid 
sns.set_style('whitegrid')
plt.figure(figsize = (10,6))
loans[loans['not.fully.paid'] == 1]['fico'].hist(alpha = 0.5,color = 'blue', 
                                              bins = 30,label = 'Not.Fully.Paid=1')
loans[loans['not.fully.paid'] == 0]['fico'].hist(alpha = 0.5,color = 'red', 
                                              bins = 30,label = 'Not.Fully.Paid=0')
plt.legend()
plt.xlabel('FICO')
```

```python
# Visualizando a contagem de empréstimos por finalidade
plt.figure(figsize = (10,6))
sns.set_style('whitegrid')
sns.countplot(x = 'purpose', hue = 'not.fully.paid', data = loans, palette = 'Set2')
```

```python
# Tendência entre o índice FICO e a taxa de juros
sns.set_style('whitegrid')
sns.jointplot(x = ['fico'], y = ['int.rate'], data = loans, color = 'green')
```

```python
# Ver se a tendência diferiu entre not.fully.paid e credit.policy
plt.figure(figsize=(12,7))
sns.lmplot(y = 'int.rate', x = 'fico', data = loans, hue = 'credit.policy',
           col = 'not.fully.paid', palette = 'Set2')
```

## <b> Padronização das Variáveis

```python
# Nova coluna
p = ['purpose']
```

```python
# Criando df com colunas das categorias
purpose = pd.get_dummies(loans, columns = p, drop_first = True)
purpose.info()
```

## <b> Divisão dos Dados em Treino e Teste

```python
# Importando o split do SKlearn
from sklearn.model_selection import train_test_split
```

```python
# Dividindo as variáveis 
X = purpose.drop('not.fully.paid',axis=1)
y = purpose['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

## <b>  Treinando um Modelo em Decision Tree

```python
# Importando o Calssificador Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
# Criando a instância do Classifier
dtree = DecisionTreeClassifier()
```

```python
# Ajustando o modelo
dtree.fit(X_train,y_train)
```

## <b> Previsões e Avaliações da Decision Tree

```python
# Predizendo os dados de teste
predictions = dtree.predict(X_test)
```

```python
# Importando métricas de avaliação do SKLearn
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# Relatório de Classificação
print(classification_report(y_test,predictions))
```

```python
# Relatório Matriz de Confusão
print(confusion_matrix(y_test,predictions))
```

## <b>  Treinando um Modelo com Random Forest

```python
# Importando o Classificador Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
# Criando a instância do modelo
rfc = RandomForestClassifier(n_estimators=600)
```

```python
# Ajustando o modelo
rfc.fit(X_train,y_train)
```

## <b> Previsões e Avaliações da Random Forest

```python
# Predizendo os dados de teste
predictions = rfc.predict(X_test)
```

```python
# Importando métricas de avaliação do SKLearn
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# Relatório de Classificação
print(classification_report(y_test,predictions))
```

```python
# Relatório Matriz de Confusão
print(confusion_matrix(y_test,predictions))
```

#### <b> Análise final:
            No modelo Decision Trees - a precisão é de que 0.85 tendem a pagar o emprestimo e 0.19 não 
            No modelo Randon Forest - a precisão é de que 0.85 tendem a pagar o emprestimo e 0.54 não 
    
    O Random Forest teve uma melhor performance, mas na análise do recall e f1-score teve uma análise inferior
    ao Decision Tree, porém a Random tende a ser melhor.
