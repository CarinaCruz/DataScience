# <b> Análise de Dados sobre Câncer com Support Vector Machines

### <b> Importando Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### <b> Obtendo Dados

```python
# Importando um conjunto de dados do SKLearn
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
# O conjunto de dados é apresentado em uma forma de dicionário:
cancer.keys()
```

```python
# Visualizando os nomes presentes no dataset
cancer['feature_names']
```

###  <b>  Informações sobre os Dados

```python
# Transformando o conjunto de dados de dicionário em um dataset
cancer_data = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
# Visualizando as informações
cancer_data.info()
```

```python
cancer_data.describe()
```

```python
# Transformando a variável target em dataframe
cancer_target = pd.DataFrame(cancer['target'],columns=['Cancer'])
cancer_target.head()
```

###  <b>  Divisão em Treino e Teste

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer_data, np.ravel(cancer_target), test_size=0.30, random_state=101)
```

```python
X_train.head()
```

```python
X_test.head()
```

###  <b>  Treinando o Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
svc = SVC(gamma='auto')
```

```python
svc.fit(X_train, y_train)
```

####  <b>  Previsões e Avaliações do SVC

```python
predictions = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

###  <b>  GridSearch

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
# Talvez demore um pouco
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

####   <b>  Previsões e Avaliações do Gridsearch

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
